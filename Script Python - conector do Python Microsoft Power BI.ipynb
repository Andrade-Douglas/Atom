```
import pandas as pd
import numpy as np
from numpy import random
from datetime import datetime

areas_imoveis = [32,50,70,90,100,150,200,250,300,350,400]
area_imovel = [random.choice(areas_imoveis) for i in range(1000)]
fatoImobiliaria = pd.DataFrame(data = area_imovel, columns=['area_imovel'])

idades_imoveis = [[1,2,3], [1,2,3,4,5], [1,2,3,4], [1,2,3,4,5,6], [1,2,3,4], [3,4,5,6,7], [3,4,5,6]]

def idade_imovel(area):
  if 32 >= area or area <= 90:
    return random.choice(idades_imoveis[0]) 
  elif area == 100:
    return random.choice(idades_imoveis[1])
  elif 150 == area or area == 200:
    return random.choice(idades_imoveis[2])
  elif area == 250:
    return random.choice(idades_imoveis[3])  
  elif area == 300:
    return random.choice(idades_imoveis[4])
  elif area == 350:
    return random.choice(idades_imoveis[5])
  elif area == 400:
    return random.choice(idades_imoveis[6])
  else:
    return 'Undefined'

fatoImobiliaria['idade_imovel'] = fatoImobiliaria['area_imovel'].map(idade_imovel)

estados_imoveis = {'id_estado_imovel':[1,2,3,4,5,6], 
                  'estado':['Rio Grande do Sul','Santa Catarina','São Paulo','Rio de Janeiro','Minas Gerais','Bahia'],
                  'uf':['RS','SC','SP','RJ','MG','BA'],
                  'pais':['Brasil','Brasil','Brasil','Brasil','Brasil','Brasil']
                  }

dimEstado = pd.DataFrame(data = estados_imoveis)

fatoImobiliaria['id_estado_imovel'] = [random.choice(dimEstado.id_estado_imovel) for i in range(1000)]

dormitorios_imoveis = [[1,2], [1,2], [2,3], [3,4], [4,5]]

def dormitorio_imovel(area):
  if 32 >= area or area <= 70:
    return random.choice(dormitorios_imoveis[0]) 
  elif area == 90:
    return random.choice(dormitorios_imoveis[1])
  elif 100 >= area or area <= 200:
    return random.choice(dormitorios_imoveis[2])
  elif area == 250:
    return random.choice(dormitorios_imoveis[3])  
  elif area >= 300 or area <= 400:
    return random.choice(dormitorios_imoveis[4])
  else:
    return 'Undefined'

fatoImobiliaria['dormitorio_imovel'] = fatoImobiliaria['area_imovel'].map(dormitorio_imovel)

piscina_imovel = {'id_piscina_imovel':[0,1], 'piscina_imovel':['Não','Sim']}
dimPiscina = pd.DataFrame(data=piscina_imovel)

piscinas_imoveis = [[0,1], [1]]

def piscina_imovel(area):
  if 32 >= area or area <= 70:
    return random.choice(piscinas_imoveis[0]) 
  elif area >= 90 or area <= 400:
    return random.choice(piscinas_imoveis[1])
  else:
    return 'Undefined'

fatoImobiliaria['id_piscina_imovel'] = fatoImobiliaria['area_imovel'].map(piscina_imovel)

elevadores_imoveis = {'id_elevador_imovel':['0','1'], 'elevador_imovel':['Não','Sim']}
dimElevador = pd.DataFrame(data=elevadores_imoveis)

elevadores_imoveis = [[0,1], [1]]

def elevador_imovel(area):
  if 32 >= area or area <= 70:
    return random.choice(piscinas_imoveis[0]) 
  elif area >= 90 or area <= 400:
    return random.choice(piscinas_imoveis[1])
  else:
    return 'Undefined'

fatoImobiliaria['id_elevador_imovel'] = fatoImobiliaria['area_imovel'].map(elevador_imovel)

garagens_imoveis = [[0,1], [1,2], [2,3], [3,4]]

def garagem_imovel(area):
  if 32 >= area or area <= 70:
    return random.choice(garagens_imoveis[0])
  elif 90 <= area <= 200:
    return random.choice(garagens_imoveis[1])
  elif 250 <= area <= 300:
    return random.choice(garagens_imoveis[2])
  elif 350 >= area or area <= 400:
    return random.choice(garagens_imoveis[3])
  else:
    return 'Undefined'

fatoImobiliaria['garagem_imovel'] = fatoImobiliaria['area_imovel'].map(garagem_imovel)

classe_bairro_imovel = {'id_classe_bairro_imovel':[1,2,3,4,5,6], 'classe_bairro_imovel':['A+','A','B','C','D','E']}
dimClasseBairro = pd.DataFrame(data=classe_bairro_imovel)

classes_bairros_imoveis = [['4','5','6'], ['3','4'], ['2','3'], ['1']]

def classes_bairros(area):
  if 32 >= area or area <= 70:
    return random.choice(classes_bairros_imoveis[0])
  elif 90 <= area <= 100:
    return random.choice(classes_bairros_imoveis[1])
  elif 150 <= area <= 250:
    return random.choice(classes_bairros_imoveis[2])
  elif 300 >= area or area <= 400:
    return random.choice(classes_bairros_imoveis[3])
  else:
    return 'Undefined'

fatoImobiliaria['id_classe_bairro_imovel'] = fatoImobiliaria['area_imovel'].map(classes_bairros)

situacoes_imoveis = {'id_situacao_imovel':[1,2,3,4], 'situacao_imovel':['Alugado-Sim','Alugado-Não','Vendido-Sim','Vendido-Não']}
dimSituacao = pd.DataFrame(data = situacoes_imoveis)

fatoImobiliaria['id_situacao_imovel'] = [random.choice(dimSituacao.id_situacao_imovel) for i in range(1000)]

idades = range(25,65)

def idade_cliente(situacao_imovel):
  if situacao_imovel == 1 or situacao_imovel == 3:
    return random.choice(idades)
  else:
    return ''

fatoImobiliaria['idade_cliente'] = fatoImobiliaria['id_situacao_imovel'].map(idade_cliente)

def data_cliente(situacao_imovel):
  if situacao_imovel == 1 or situacao_imovel == 3:
    return random.choice(pd.date_range(start='1-1-2015', end='1-7-2020'))
  else:
    return ''

fatoImobiliaria['data_registro'] = fatoImobiliaria['id_situacao_imovel'].map(data_cliente)

area_32_50 = range(80000,150000)
area_70 = range(150000,220000)
area_90 = range(220000,350000)
area_100 = range(300000,450000)
area_150 = range(450000,600000)
area_200 = range(600000,750000)
area_250 = range(700000,900000)
area_300 = range(1000000,1500000)
area_350 = range(1600000,2000000)
area_400 = range(2500000,3000000)

def valor_venda(area):
  if 32 >= area or area <= 50:
    return random.choice(area_32_50)
  elif area == 70:
    return random.choice(area_70)
  elif area == 90:
    return random.choice(area_90)
  elif area == 100:
    return random.choice(area_100)
  elif area == 150:
    return random.choice(area_150)
  elif area == 200:
    return random.choice(area_200)
  elif area == 250:
    return random.choice(area_250)
  elif area == 300:
    return random.choice(area_300)
  elif area == 350:
    return random.choice(area_350)
  elif area == 400:
    return random.choice(area_400)
  
  else:
    return 'Undefined'

fatoImobiliaria['valor_venda_imovel'] = fatoImobiliaria['area_imovel'].map(valor_venda)

area_32_50 = range(800,1500)
area_70 = range(1500,2200)
area_90 = range(2200,3500)
area_100 = range(3000,4500)
area_150 = range(4500,6000)
area_200 = range(6000,7500)
area_250 = range(7000,9000)
area_300 = range(10000,15000)
area_350 = range(16000,20000)
area_400 = range(25000,30000)

def valor_aluguel(area):
  if 32 >= area or area <= 50:
    return random.choice(area_32_50)
  elif area == 70:
    return random.choice(area_70)
  elif area == 90:
    return random.choice(area_90)
  elif area == 100:
    return random.choice(area_100)
  elif area == 150:
    return random.choice(area_150)
  elif area == 200:
    return random.choice(area_200)
  elif area == 250:
    return random.choice(area_250)
  elif area == 300:
    return random.choice(area_300)
  elif area == 350:
    return random.choice(area_350)
  elif area == 400:
    return random.choice(area_400)
  
  else:
    return 'Undefined'

fatoImobiliaria['valor_aluguel_imovel'] = fatoImobiliaria['area_imovel'].map(valor_aluguel)

fatoImobiliaria	
```
